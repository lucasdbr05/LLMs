# Embeding

```python
from langchain_ollama import OllamaEmbeddings

embedding_model = OllamaEmbeddings(
    model="llama2",
)
```

```python
embedings = embedding_model.embed_documents(
    [
        'Eu gosto de cochorros',
        'Eu gosto de animais',
        'O tempo est ruim l fora'
    ]
)
embedings
```

```python
import numpy as np
```

```python
for i in range(len(embedings)):
    for j in range(len(embedings)):
        print(np.dot(embedings[i], embedings[j]), end=' | ')
    print()
```

### Embed query

```python
pergunta = 'O que  um cachorro'
emb_query = embedding_model.embed_query(pergunta)
```
